```python
#4 What block has the most and does that mean its the most dangerous? (not including unknowns or homeless)
MBLOCK = Offenders.groupby("BLOCK")["LAST"].count()
large5 = MBLOCK.nlargest(5)
large5
```

```python
#5 What percentage of sex offenders were registered as homeless? 

MBLOCK = Offenders.groupby("BLOCK")["BLOCK"].count()
homeless = MBLOCK.nlargest(1)
homeless.sum()

```

```python
# 8.	What year has the highest number of registered sex offenders? How many? – Medium
BIGYEAR = Offenders.groupby("YEAR", sort=False)["YEAR"].count()
large1 = BIGYEAR.nlargest(1)
large1
```
